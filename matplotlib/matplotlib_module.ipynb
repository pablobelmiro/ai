```python
import numpy as np
import matplotlib.pyplot as plt
```

## Basic functions

```python
x = np.linspace(0, 5, 11)
y = x ** 2

%matplotlib inline
plt.plot(x, y)
#chart settings
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("My first chart")
```

```python
%matplotlib inline

#Ploting 2 axis at one figure
plt.subplot(1, 2, 1)
plt.plot(x, y, "r--")

plt.subplot(1, 2, 2)
plt.plot(y, x, "g*-")
```

```python
%matplotlib inline

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

#using subplots(Docs example)
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```
