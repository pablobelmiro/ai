```python
import numpy as np
import matplotlib.pyplot as plt
```

## Subplots

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
#Figures are tuples!
fig, ax = plt.subplots() #Desestructuring a tuple
print(fig)
print(ax)

#print chart
%matplotlib inline
ax.plot(x, y, "r--")
```

```python
#creating figure with settings
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

#List of axis (2charts)
for axis in ax:
    print(axis)
    
#manipulations are made at individual axis
for axis in ax:
    axis.plot(x, y, "r--")
```

```python
#working with list of lists (charts)
fig, ax = plt.subplots(nrows=2, ncols=2)

#running on rows
for axis in ax:
    #running on columns
    for chart in axis:
        print(chart)
```
