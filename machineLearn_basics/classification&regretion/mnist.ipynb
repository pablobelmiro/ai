## Mnist Scikit-learn

```python
import matplotlib.pyplot as mpl
import numpy as np

from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('mnist_784', version=1, as_frame=False)

#Accessing the keys(like a dictionary) of the dataset
mnist.keys()

x, y = mnist["data"], mnist["target"]

#using numpy to cast data type to float
y = y.astype(float)
```

```python
numberImgsToShow = 10

fig, axes = mpl.subplots(1, numberImgsToShow, figsize=(20, 2))
for i in range(10):
    ax = axes[i]
    ax.imshow(x[i].reshape(28, 28), cmap="binary")
    ax.axis("off")
mpl.show()
```
